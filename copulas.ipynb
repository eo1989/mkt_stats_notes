```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
import numpy as np
```

```python
_x = st.uniform(0, 1).rvs(10_000)
sns.displot(_x)
```

```python
norm = st.distributions.norm()
x_trans = norm.ppf(_x)
sns.histplot(x_trans)
```

```python
# _h = sns.jointplot(x = _x, y = x_trans)
# _h = sns.
# _h.set_axis_labels("Original", "Transformed", fontsize=16)


def qqplot(x, y, **kwargs):
    _, xr = st.probplot(_x, fit=False)
    _, yr = st.probplot(x_trans, fit=False)
    plt.scatter(xr, yr, **kwargs)


g = sns.FacetGrid(data=_x, col="x")
g.map(qqplot, "observed", "transformed")
```
